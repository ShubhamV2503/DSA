#### Method Overriding

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

animals = [Dog(), Cat()]

for animal in animals:
    animal.speak()
```

```python
class Shape:
    def area(self):
        return "Area not defined for generic shape"

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
    
class triangle(Shape):
    def __init__(self , base, height):
        self.base = base
        self.height = height

# List of shapes
shapes = [Circle(5), Rectangle(4, 6) , triangle(10,20)]

# Polymorphism in action
for shape in shapes:
    print(f"The area is: {shape.area()}")
```

```python
# Shape is an abstract base class (ABC) that defines a c
# ommon interface for all shapes.
# The @abstractmethod decorator ensures that any subclass 
# of Shape must implement the area() method. 
# If a subclass does not override area(), 
# Python will raise an error when you try to instantiate it.



from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        return "Area not defined for generic shape"

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
    
class triangle(Shape):
    def __init__(self , base, height):
        self.base = base
        self.height = height

# List of shapes
shapes = [Circle(5), Rectangle(4, 6) , triangle(10,20)]

# Polymorphism in action
for shape in shapes:
    print(f"The area is: {shape.area()}")
```

```python
# 2. E-commerce Checkout:

# Write a function checkout(item) that accepts either Book, Clothing, 
# or Gadget objects.

# Each class should have a method price() that returns a 
# different calculation



class checkout:
    def price(self):
        return 'Not Found Price Method for this Option'

class book(checkout):
    def __init__(self,qty,amt):
        self.qty = qty
        self.amt = amt

    def price(self):
        return f'Total Checkout Amt of Book is {self.qty * self.amt}'


class cloth(checkout):
    def __init__(self,qty,amt):
        self.qty = qty
        self.amt = amt

    def price(self):
        return f'Total Checkout Amt of cloth is {self.qty * self.amt}'

class gadget(checkout):
    def __init__(self,qty,amt):
        self.qty = qty
        self.amt = amt

    def price(self):
        return f'Total Checkout Amt of gadget is {self.qty * self.amt}'


b = book(2,10)
c = cloth(3,30)
g = gadget(4,40)

print(b.price())
print(c.price())
print(g.price())
```

```python
class CreditCard:
    def pay(self, amount):
        print(f"Paid {amount} via Credit Card")

class UPI:
    def __init__(self, upi_id):
        self.upi_id = upi_id

    def pay(self, amount):
        masked = self.upi_id[0] + "***" + self.upi_id[self.upi_id.find('@'):]
        print(f"Paid {amount} via UPI ({masked})")

class PayPal:
    def pay(self, amount):
        if amount > 10000:
            print("PayPal: Transaction declined (limit exceeded)")
        else:
            print(f"Paid {amount} via PayPal")

def checkout(cart_total, payment_method):
    payment_method.pay(cart_total)

# Test
checkout(3000, CreditCard())
checkout(1500, UPI("user@upi"))
checkout(12000, PayPal())
```
