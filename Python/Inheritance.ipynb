### Single Inheritance

```python
class User:
    def __init__(self, username):
        self.username = username
        print(self.username)

    def login(self):
        print(f"{self.username} has logged in.")

class AdminUser(User):
    def delete_user(self, userid):
        print(f"{self.username} deleted user: {userid}")

# Usage
admin = AdminUser("superadmin")
admin.login()
admin.delete_user("user123")


#📌 Use Case: System admin inherits generic user behavior and adds admin privileges.
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def display_info(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")

class IssuedBook(Book):
    def __init__(self, title, author, issued_to, due_date):
        super().__init__(title, author)  # Call the parent constructor
        self.issued_to = issued_to
        self.due_date = due_date

    def display_issued_info(self):
        # You can call parent method if you want to show book info as part of issued info
        self.display_info()
        print(f"Issued To: {self.issued_to}")
        print(f"Due Date: {self.due_date}")

# 🔍 Test the implementation
b1 = IssuedBook("The Alchemist", "Paulo Coelho", "John Doe", "2025-09-10")
b1.display_issued_info()
```

```python
class Company:

    def company_name(self,company):
        self.company = company
        print(f"Company name is : {self.company}")

class Customer(Company):

    def __init__(self,name,email):
        self.name = name
        self.email = email

    def display_customer_info(self):
        print(f"customer name is : {self.name}")
        print(f"Email is : {self.email}")
        super().company_name('Google')

    
class Invoice(Customer):

    def __init__(self, name , email, invoice_number, amount):
        super().__init__(name,email) ## Inheriting parent class constructor
        self.invoice_number = invoice_number
        self.amount = amount

    def generate_invoice(self):
        print(f"invoice number : {self.invoice_number}")
        print(f"amount : {self.amount}")

c1 = Invoice('Joyboy', 'joyboy2503@outlook.com', 'INV-2025', 10000)
c1.display_customer_info()
c1.generate_invoice()
c1.company_name('Microsoft')
print("************"*4)


c2 = Customer('Shubham Vishwakarma','vishwakarmashubham.2503@outlook.com')
c2.display_customer_info()
c2.company_name('Apple')
```

### Multiple Inheritance

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def get_balance(self):
        return self.balance + 100

class KYCVerified:
    def verify_kyc(self):
        print("KYC documents approved.")

class VerifiedAccount(BankAccount, KYCVerified):
    def access_international_features(self):
        print("Accessing international transfers...")

# Usage
va = VerifiedAccount(10000)
print(va.get_balance())
va.verify_kyc()
va.access_international_features()
```
