```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance or invalid amount")

    def get_balance(self):
        return self.__balance
```

```python
object = BankAccount(100)
object.deposit(200)
print('current balance is = ',object.get_balance())
```

```python
# Q1: Student Marks Manager – Solution

class Student:
    def __init__(self):
        self.__marks = 0  # Private variable

    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Invalid marks! Please enter between 0 and 100.")

    def get_marks(self):
        return self.__marks

# 🔍 Test the class
s1 = Student()
s1.set_marks(85)
print("Marks:", s1.get_marks())  # ✅ Output: Marks: 85

s1.set_marks(120)                # ❌ Invalid
print("Marks:", s1.get_marks())  # ✅ Still: 85 (unchanged)
```

```python

##  Protected Attributes and Methods

class Car:
  def __init__(self, model):
    self._model = model # protected attribute

  def _get_info(self): # protected method
    return f"Model: {self._model}"

class SportsCar(Car):
  def show_info(self):
    return self._get_info()

car = Car('Toyota')
sports = SportsCar('Ferrari')
print(car._model)
print(sports.show_info())
```
